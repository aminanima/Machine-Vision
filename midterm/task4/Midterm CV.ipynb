```python

import numpy as np
import matplotlib.pyplot as plt
import scipy.misc as img
from scipy import ndimage as nimg
import random
from skimage import feature

im = img.imread("lena_cover.jpg", mode = 'F')

image = feature.canny(im, sigma = 5)
image2 = feature.canny(im, sigma = 20)

image3 = nimg.sobel(im)

fig = plt.figure(figsize = (10, 15))

plt.subplot(331)
plt.imshow(im, cmap='gray')
plt.subplot(332)
plt.imshow(image, cmap='Reds')
plt.subplot(333)
plt.imshow(image2, cmap='Reds')
plt.subplot(334)
plt.imshow(image3 ,cmap='Reds')
plt.show()
```
