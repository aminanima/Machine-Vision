```python
#TASK2
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
import numpy as np
import matplotlib.image as img
im = pltimg.imread('lena_cover.jpg')
fig = plt.figure()
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.114])

img = pltimg.imread('lena_cover.jpg')     
gray = rgb2gray(img)    

fig = plt.figure()

plt.subplot(222)
plt.imshow(im)
plt.subplot(223)
plt.imshow(gray, cmap = plt.get_cmap('gray'))
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as img
%matplotlib inline
imlena = img.imread('./lena_cover.jpg')
from scipy import ndimage
import scipy.misc
import numpy as np
import math
from scipy import ndimage as nimg
import numpy as np
from scipy.misc.pilutil import Image
a = Image.open('./lena_cover.jpg')
b = scipy.misc.fromimage(a)
b1 = b.astype(float)
b2 = np.max(b1)
c = (255.0*np.log(70+b1))/np.log(b2)
c1 = c.astype(int)
d = scipy.misc.toimage(c1)
fig = plt.figure(figsize=(15,10))
plt.subplot(221)
plt.imshow(imlena)
plt.subplot(222)
plt.imshow(d)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.misc as img
from scipy import ndimage as nimg
import cv2

im = img.imread("lena_cover.jpg")
img = cv2.imread('lena_cover.jpg',0)
hist = cv2.calcHist([img],[0],None,[256],[0,256])
startIm = im
fig = plt.figure(figsize = (10, 15))
plt.subplot(331)
plt.imshow(startIm)

sumRed = 0
sumGreen = 0
sumBlue = 0
for i in range (im.shape[0]):
    for j in range(im.shape[1]):
        sumRed += im[i, j, 0]
        sumGreen += im[i, j, 1]
        sumBlue += im[i, j, 2]
        
mRed = sumRed / im.size
mGreen = sumGreen / im.size
mBlue = sumBlue / im.size

avg = (mRed + mGreen + mBlue) / 3

for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        im[i, j, 0] = im[i, j, 0] * avg / mRed
        im[i, j, 1] = im[i, j, 1] * avg / mGreen
        im[i, j, 2] = im[i, j, 2] * avg / mBlue

        
plt.subplot(332)
plt.plot(hist)        
plt.subplot(333)
plt.imshow(im)
plt.show()
```
